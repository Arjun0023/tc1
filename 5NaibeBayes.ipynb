```python
import pandas as pd
```

```python
iris = pd.read_csv('Iris.csv')
```

```python
iris.head(10)
```

```python
x = iris[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]   #independent attribute
y = iris[['Species']]                                                       #dependent attribute
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar = MinMaxScaler()
x_scaled = scalar.fit_transform(x)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled,y,test_size=0.20,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
gnb.fit(x_train,y_train)
```

```python
y_pred = gnb.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
import sklearn.metrics
```

```python
lbs = ['Iris-versicolor','Iris-setosa','Iris-virginica']
```

```python
print(sklearn.metrics.confusion_matrix(y_test,y_pred,labels= lbs))
```
